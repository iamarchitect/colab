```python
import torch
from transformers import GPT2Tokenizer, GPT2LMHeadModel

# Load the fine-tuned GPT-2 model
model_path = "/content/drive/MyDrive/fin_gpt_result/finetuned_gpt2_instruction_model"
model = GPT2LMHeadModel.from_pretrained(model_path)
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
tokenizer.add_special_tokens({'pad_token': '<pad>'})  # Set '<pad>' as the padding token

# Set the device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Set the maximum length for generated text
max_length = 900

input_data = [
    {"instruction": "Given below the balance sheet of HDFC Bank. Provide a financial insight based on below data.\n"
            "Mar 2020\n"
            "Share Capital: 548.0\n"
            "Reserves: 175810.0\n"
            "Borrowings: 1333041.0\n"
            "Other Liabilities: 71430.0\n"
            "Fixed Assets: 4776.0\n"
            "CWIP: 0.0\n"
            "Investments: 389305.0\n"
            "Other Assets: 1186750.0\n"

            "Mar 2021\n"
            "Share Capital: 551.0\n"
            "Reserves: 209259.0\n"
            "Borrowings: 1511418.0\n"
            "Other Liabilities: 78279.0\n"
            "Fixed Assets: 5248.0\n"
            "CWIP: 0.0\n"
            "Investments: 438823.0\n"
            "Other Assets: 1355435.0\n"

            "Mar 2022\n"
            "Share Capital: 555.0\n"
            "Reserves: 246772.0\n"
            "Borrowings: 1784970.0\n"
            "Other Liabilities: 90639.0\n"
            "Fixed Assets: 6432.0\n"
            "CWIP: 0.0\n"
            "Investments: 449264.0\n"
            "Other Assets: 1667238.0\n",
     "text": "Over the past three years, HDFC Bank has exhibited positive financial trends and growth.From Mar 2020 to Mar 2022, the bank has shown a consistent increase in share capital, reserves, borrowings, and other liabilities. This indicates the bank's ability to raise capital, retain earnings, and attract external funding. HDFC Bank has made substantial investments, with the value of investments growing from 389,305 in Mar 2020 to 449,264 in Mar 2022. The bank has also expanded its asset base, as seen in the growth of fixed assets and other assets. These trends suggest that HDFC Bank has been actively investing in infrastructure and other assets to support its operations and growth strategy. The growth in borrowings and other liabilities indicates the bank's increasing capacity to take on additional obligations. Overall, these financial insights demonstrate HDFC Bank's strong financial position and its efforts to expand its asset base and fund its operations through a combination of capital, reserves, and borrowings."}
]

# Set the input instruction
input_instruction = input_data[0]["instruction"]

# Encode the instruction
input_tokens = tokenizer.encode(input_instruction, add_special_tokens=True)
input_tensor = torch.tensor([input_tokens], dtype=torch.long).to(device)
attention_mask = torch.ones_like(input_tensor)
# Generate text based on the instruction
model.eval()
with torch.no_grad():
    output = model.generate(input_tensor,attention_mask=attention_mask,
                            max_length=max_length, num_return_sequences=1,
                            pad_token_id=tokenizer.pad_token_id)
    generated_tokens = output[0].tolist()
    generated_text = tokenizer.decode(generated_tokens[len(input_tokens):], skip_special_tokens=True)

# Print the generated text
print("Generated Text:", generated_text)
```
